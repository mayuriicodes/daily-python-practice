```python
#print table of number using while
i=1
while(i<=10):
    print(i*2)
    i=i+1
```

```python
#reverse number
i=10
while(i>=1):
    print(i)
    i=i-1
print('done')
```

```python
#wap calculate factorial of given number
fact=5
res=1
i=1
while (i<=fact):
    res=res*i
    i=i+1
print(res)
```

```python
#convert into  capital
name=input('enter name : ')
for i in name:
    if i in 'abcdefghijklmnopqrstuvwxyz':
        print(i.upper(),end='')
```

```python
#wap to print triangle of *
for i in range (5):
    print('*'*i)
```

```python
#wap to check number is prime or not 
num=int(input('enter number'))
for i in range (2,num):
    if num%i==0:
        print('not prime')
        break
else:
    print('prime')
```

```python
for i in range(5):
    if i%2==1:       #if odd print '&'
        print('&'*i)
    else:            #if even print '#'
        print('#'*i)
```
